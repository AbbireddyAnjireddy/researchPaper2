```python
!pip install xlsxwriter
import pandas as pd
import re
import os
import xlsxwriter

# === STEP 1: Define your CSV paths ===
csv_files = [
    "/content/all_csvs/archive/CEAS_08.csv",
    "/content/all_csvs/archive/SpamAssasin.csv",
    "/content/all_csvs/archive/Nigerian_Fraud.csv",
     "/content/all_csvs/archive/Nazario.csv"
]
# === STEP 2: Domain keywords and colors ===
domain_keywords = {
    'finance': ['bank', 'account', 'transaction', 'credit', 'debit', 'payment', 'loan', 'investment', 'fund'],
    'retail': ['order', 'invoice', 'delivery', 'cart', 'item', 'purchase', 'receipt', 'store', 'sale'],
    'hospital': ['doctor', 'medical', 'hospital', 'patient', 'treatment', 'prescription', 'healthcare', 'clinic']
}
domain_colors = {
    'finance': '#0070C0',
    'retail': '#C00000',
    'hospital': '#00B050'
}

# === STEP 3: Patterns to remove noisy log/debug messages ===
log_patterns = [
    r"dbg:", r"sa-update", r"Net::DNS", r"SpamAssassin", r"channel updates\.",
    r"perl platform", r"module installed", r"logger:", r"score set \d", r"/var/lib/spamassassin"
]
combined_pattern = re.compile("|".join(log_patterns), re.IGNORECASE)

def is_clean(text):
    if not isinstance(text, str):
        return True
    return not combined_pattern.search(text)

def build_rich_text(text, matched_keywords, default_format, highlight_formats):
    if not isinstance(text, str):
        return None, str(text)
    parts, last_idx = [], 0
    pattern = r'\b(' + '|'.join(re.escape(k) for ks in domain_keywords.values() for k in ks) + r')\b'
    for match in re.finditer(pattern, text, re.IGNORECASE):
        start, end = match.span()
        word = match.group(0)
        domain = next((d for d, kws in domain_keywords.items() if word.lower() in kws), None)
        if start > last_idx:
            parts += [default_format, text[last_idx:start]]
        parts += [highlight_formats.get(domain, default_format), text[start:end]]
        matched_keywords.add(word.lower())
        last_idx = end
    if last_idx < len(text):
        parts += [default_format, text[last_idx:]]
    return (parts if len(parts) >= 3 else None), text

# === STEP 4: Output folder ===
output_folder = "./processed_excel/"
os.makedirs(output_folder, exist_ok=True)

# === STEP 5: Process each CSV file ===
for file_path in csv_files:
    base_name = os.path.splitext(os.path.basename(file_path))[0]
    output_path = os.path.join(output_folder, f"{base_name}_highlighted.xlsx")

    try:
        df = pd.read_csv(file_path)
    except Exception as e:
        print(f"‚ùå Error reading {file_path}: {e}")
        continue

    # Clean: drop rows with missing essential columns and noisy logs
    required_columns = ['subject', 'body', 'sender', 'receiver', 'date']
    df = df.dropna(subset=required_columns).drop_duplicates()
    df = df[df['body'].apply(is_clean) & df['subject'].apply(is_clean)]

    wb = xlsxwriter.Workbook(output_path)
    ws = wb.add_worksheet("Emails")

    # Formatting
    highlight_formats = {d: wb.add_format({'color': c}) for d, c in domain_colors.items()}
    default_format = wb.add_format({'color': 'black', 'text_wrap': True})

    headers = list(df.columns) + ['matched_domain', 'matched_keyword']
    for col_idx, col in enumerate(headers):
        ws.write(0, col_idx, col)

    # Row-wise writing
    for row_idx, row in df.iterrows():
        matched_keywords = set()
        matched_domain = ""

        for col_idx, col in enumerate(df.columns):
            value = str(row[col]) if pd.notna(row[col]) else ""
            if col.lower() in ['subject', 'body']:
                parts, fallback = build_rich_text(value, matched_keywords, default_format, highlight_formats)
                if parts:
                    ws.write_rich_string(row_idx + 1, col_idx, *parts)
                else:
                    ws.write(row_idx + 1, col_idx, fallback)
                if not matched_domain:
                    for domain, kws in domain_keywords.items():
                        if any(kw in value.lower() for kw in kws):
                            matched_domain = domain
                            break
            else:
                ws.write(row_idx + 1, col_idx, value)

        ws.write(row_idx + 1, len(df.columns), matched_domain)
        ws.write(row_idx + 1, len(df.columns) + 1, ", ".join(sorted(matched_keywords)))

    # Column formatting
    for i, col in enumerate(headers):
        if col.lower() in ['subject', 'body']:
            ws.set_column(i, i, 60)
        elif col.lower() in ['matched_domain', 'matched_keyword']:
            ws.set_column(i, i, 30)
        else:
            ws.set_column(i, i, 20)

    wb.close()
    print(f"‚úÖ Saved: {output_path}")
```

```python
import pandas as pd

# ‚úÖ Paths to your highlighted Excel files
highlighted_files = {
    "CEAS_08": "/content/processed_excel/CEAS_08_highlighted.xlsx",
    "SpamAssassin": "/content/processed_excel/SpamAssasin_highlighted.xlsx",
    "Nigerian_Fraud": "/content/processed_excel/Nigerian_Fraud_highlighted.xlsx",
    "Nazario": "/content/processed_excel/Nazario_highlighted.xlsx"
}

# ‚úÖ Required columns for cleaning
required_columns = ['subject', 'body', 'sender', 'receiver', 'date',
                    'matched_domain', 'matched_keyword', 'label']

# ‚úÖ Store all dataset summaries for final merge
combined_summary = []

# ‚úÖ Loop through each file and print individual summary
for name, path in highlighted_files.items():
    try:
        df = pd.read_excel(path)
        df_cleaned = df.dropna(subset=required_columns).drop_duplicates()

        domain_counts = df_cleaned['matched_domain'].value_counts()
        label_counts = df_cleaned.groupby('matched_domain')['label'].value_counts().unstack().fillna(0).astype(int)

        summary = pd.DataFrame({
            'Total Emails': domain_counts,
            'Phishing (1)': label_counts.get(1, 0),
            'Legitimate (0)': label_counts.get(0, 0)
        }).astype(int)

        summary.loc['Total'] = summary.sum()

        print(f"\nüìÇ Dataset: {name}")
        print(summary.to_markdown(tablefmt="grid"))

        # Store with dataset name
        summary["Dataset"] = name
        combined_summary.append(summary.reset_index())

    except Exception as e:
        print(f"‚ùå Error processing {name} - {path}: {e}")

# ‚úÖ Combine all into a single summary table
if combined_summary:
    final_summary = pd.concat(combined_summary, ignore_index=True)
    final_summary = final_summary.rename(columns={"index": "Domain"})

    print("\nüìä üì¶ Combined Summary Table (All Datasets):")
    print(final_summary.to_markdown(tablefmt="grid"))
```

```python
import pandas as pd

# ‚úÖ Step 1: Provide paths to your highlighted Excel files
highlighted_files = [
    "/content/processed_excel/CEAS_08_highlighted.xlsx",
    "/content/processed_excel/SpamAssasin_highlighted.xlsx",
    "/content/processed_excel/Nigerian_Fraud_highlighted.xlsx",
    "/content/processed_excel/Nazario_highlighted.xlsx"  # change if needed
]

# ‚úÖ Step 2: Combine cleaned data from all files
all_cleaned_data = []
required_columns = ['subject', 'body', 'sender', 'receiver', 'date',
                    'matched_domain', 'matched_keyword', 'label']

for file in highlighted_files:
    try:
        df = pd.read_excel(file)
        df_cleaned = df.dropna(subset=required_columns).drop_duplicates()
        all_cleaned_data.append(df_cleaned)
    except Exception as e:
        print(f"‚ùå Error processing {file}: {e}")

# ‚úÖ Step 3: Concatenate all cleaned data
final_df = pd.concat(all_cleaned_data, ignore_index=True)

# ‚úÖ Step 4: Count emails per domain
domain_counts = final_df['matched_domain'].value_counts()

# ‚úÖ Step 5: Phishing vs Legitimate per domain
label_counts = final_df.groupby('matched_domain')['label'].value_counts().unstack().fillna(0).astype(int)

# ‚úÖ Step 6: Build summary
summary = pd.DataFrame({
    'Total Emails': domain_counts,
    'Phishing (1)': label_counts.get(1, 0),
    'Legitimate (0)': label_counts.get(0, 0)
}).astype(int)

# ‚úÖ Step 7: Add total row
summary.loc['Total'] = summary.sum()

# ‚úÖ Step 8: Show summary table
print("üìä Cleaned Domain-Wise Email Summary (All Datasets):")
print(summary.to_markdown(tablefmt="grid"))
```
